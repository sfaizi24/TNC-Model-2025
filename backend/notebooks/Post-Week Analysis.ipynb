# Post-Week Analysis

This notebook connects to the Neon database and provides a simple `query()` function to execute SQL queries. Results are returned as pandas DataFrames.


```python
import os
from dotenv import load_dotenv
import psycopg2
import pandas as pd

# Load environment variables
load_dotenv()

# Get database connection string
DATABASE_URL = os.getenv('NEON_DATABASE_URL')

# Create connection
conn = psycopg2.connect(DATABASE_URL)

def query(sql):
    """Execute SQL query and return results as pandas DataFrame"""
    return pd.read_sql_query(sql, conn)
```

```python
query("""
    SELECT b.*, u.first_name, u.last_name
    FROM bets b
    JOIN users u ON b.user_id = u.id
    WHERE u.first_name = 'Umar' AND b.week = 11 
    ORDER BY b.result DESC;
""")
```
